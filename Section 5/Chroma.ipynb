#### Chroma

Chroma is a AI-native open source vector database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0

```python
## Building a sample vector db
from langchain_community.document_loaders import TextLoader
from langchain_chroma import Chroma
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
# Load the data
loader = TextLoader("Linkedin_About.txt")
data = loader.load()
data
```

```python
# Split the data into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_splitter.split_documents(data)
splits
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")
db = Chroma.from_documents(splits, embeddings)
db
```

```python
query = "What can you tell me about the user's profile?"
docs = db.similarity_search(query)
docs[0].page_content
```

```python
docs
```

```python
# Saving to the disk
db1 = Chroma.from_documents(documents=splits, embedding=embeddings, persist_directory='./chroma_db')
```

```python
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
docs = db2.similarity_search(query)
docs
```

```python
## similarity search with score
docs = db2.similarity_search_with_score(query)
docs
```

```python
#### Retreiver option
retriever = db.as_retriever()
retriever.invoke(query)[0].page_content
```
