```python
## Web basd loader
import bs4
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(web_path=("https://medium.com/data-and-beyond/document-loaders-in-langchain-f23d3ce70d66",),
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                           class_=("title") # Write classes whose content you need
                       ))
                       )
loader.load()
```
