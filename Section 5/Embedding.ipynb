```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is a tutorial on OpenAI Embeddings"
query_result = embeddings.embed_query(text)
query_result
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1025)
```

```python
text = "This is a tutorial on OpenAI Embeddings"
query_result = embeddings_1024.embed_query(text)
query_result
```

```python
## Text loader

from langchain_community.document_loaders import TextLoader

loader = TextLoader('Linkedin_About.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
# final_docs = text_splitter.create_documents(docs) # This works when the type of input we give is of text type and not docs

# For already created documents where we just want to split it, we use this instead
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
## Vector Embeddings and Vector StoreDB

from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_docs, embeddings_1024)
db
```

```python
query = "I worked at Micron Technology, where I built"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
