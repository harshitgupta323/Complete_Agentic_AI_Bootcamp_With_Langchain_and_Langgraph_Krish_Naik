#### How to split by Character - Character text splitter

This is the simplest method. This splits based on a given character sequence, which defaults to "\n\n". Chunk size is measured by the number of characters.

1. How the text is split: by single character sequence.
2. How the chunk size is measured: by number of characters.

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('Linkedin_About.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=100, chunk_overlap = 20)
text = text_splitter.split_documents(docs)
```
