 ## Encapsulation and Abstraction

 Encapsulation and Abstraction are two fundamental principles of Object Oriented Programming that help in designing robust, maintainable and reusable code. Encapsulation involves bundling data and methods that operate on the data in a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features.

Encapsulation is the concept of wrapping data and methods that operate on the data in a single unit. It rescricts direct access to some of the object's components, which is a powerful means of preventing accidental inference and misuse of the data.

```python
class Person:

    def __init__(self, name, age):
        self.name = name # public variable
        self.age = age # public variable

person = Person("Harshit", 30)
print(person.name)
```

```python
class Person:

    def __init__(self, name, age, gender):
        self.name = name # public variable
        self.age = age # public variable
        self.__gender = gender # private variable and cannot be accessed directly

    # Private variables can be accessed only when exposed by a method defined by the programmer
    def getGender(self):
        return self.__gender

person = Person("Harshit", 30, "Male")
print(person.name)
# print(person.gender)
# print(person.__gender)
person.getGender()
```

```python
# Protected variables can be accessed in the derived class

class Person:

    def __init__(self, name, age, gender):
        self._name = name # protected variable
        self._age = age # protected variable
        self.__gender = gender # private variable and cannot be accessed directly

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)

employee = Employee("Harshit", 30, "Male")
print(employee._name)
```
